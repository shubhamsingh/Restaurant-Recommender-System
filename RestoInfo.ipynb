```python
# Import libraries here
# import numpy as np
# from sklearn import linear_model
import pandas as pd
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import seaborn as sns

# [Example] Read datasets
# iris_train = pd.read_csv("/data/training/RestoInfo.csv")
# print(iris_train)
# User can save any number of plots to `/code/output` directory
# Note that only the plots saved in /code/output will be visible to you
# So make sure you save each plot as shown below
# Uncomment the below 3 lines and give it a try
# axes = pd.tools.plotting.scatter_matrix(iris_train, alpha=0.2)
# plt.tight_layout()
# plt.savefig('/code/output/scatter_matrix.png')

# Write output file
# Assuming iris_pred is DataFrame in the required output format
# iris_pred.to_csv("/code/iris_prediction.csv")


#read data
resto = pd.read_csv("/data/training/RestoInfo.csv")
print(resto.head(10))

#Info of data
print(resto.info())

#Description of data
print(resto.describe())

#check the null value
print(resto.isnull().sum())

print(resto.columns)
#data cleaning
"""
Analysis 1:
Removed few columns because of not useful in prydiction
Unnamed: 0: random varibale containg no information
dish_liked : most values are NUll
reviews_list : redundant feature with rate and vote
menu_item : redundant feature with rest_type, cuisines, listed_in(type) and dish_liked
"""
print("Analysis 1:")
print("Removed few columns because of not useful in prydiction")
print("Unnamed: 0: random varibale containg no information")
print("dish_liked : most values are NUll")
print("reviews_list : redundant feature with rate and vote")
print("menu_item : redundant feature with rest_type, cuisines, listed_in(type) and dish_liked")

restoX = resto[['name', 'online_order', 'book_table', 'rate', 'votes',
       'location', 'rest_type', 'cuisines', 'approx_cost(for two people)',
       'listed_in(type)', 'listed_in(city)']]

"""
Analysis 2:
Redundent feature
location: location and listed_in(city) are redundant. 
So taken listed_in(city) has super_set of location
"""
print("Analysis 2:")
print("Redundent feature")
print("location: location and listed_in(city) are redundant.")
print("So taken listed_in(city) has super_set of location")


print(restoX['listed_in(city)'].value_counts())
print(restoX['location'].value_counts())
print(restoX['listed_in(type)'].value_counts())

del restoX['location']

#data pre_processing
"""
Analysis 3:
convert rating and removed /5 which is not usefull in analysis
"""
print("Analysis 3:")
print("convert rating and removed /5 which is not usefull in analysis")

restoX['rate'] = restoX['rate'].apply(lambda x: str(x).split('/')[0])


# figure shows number of restaurtants that take online_order
sns.catplot(data = restoX, kind = 'count', x = 'online_order')
plt.title('Number of restaurants take online order')
plt.savefig('/code/output/restaurants_take_online_order.png')
#Analysis 4: 60% of orders are online in Bangalore.
print("Analysis 4:")
print("60% of orders are online in Bangalore.")


# figure shows number of restaurants in each city
lable = sns.catplot(data = restoX, kind = 'count', x = 'listed_in(city)')
plt.title('Number of restaurants in each city')
lable.set_xticklabels(rotation = 90)
plt.savefig('/code/output/restaurants_in_each_city.png')
#Analysis 5: BTM has maximum restaurants in this city with count 154
print("Analysis 5:")
print("BTM has maximum restaurants in this city with count 154.")

# figure shows number of restaurants with meal type
sns.catplot(data = restoX, kind = 'count', x = 'listed_in(type)')
plt.title('Number of restaurants according to meal type')
plt.savefig('/code/output/count_restaurants_based_on_meal_type.png')
"""
Analysis 6: 
Around 50% customer prefer to take delivery of food and
around 35% customer prefer Dine-out
"""
print("Analysis 6:")
print("Around 50% customer prefer to take delivery of food and "
"around 35% customer prefer Dine-out")

print(restoX['rate'].unique())
#Analysis 7: data contain new, -, nan which is not interpretable and replace with Average value_counts
print("Analysis 7:")
print("data contain new, -, nan which is not interpretable and "
"replace with Average value_counts")

rate = list(restoX['rate'])
try:
    for i in range(0, len(rate)):
        if rate[i] == 'nan' or rate[i] == '-' or rate[i] == 'NEW':
            rate[i] = 2.5 #assign average value
        else:
            rate[i] = float(rate[i])
except:
    pass
restoX['rate'] = rate 

# figure shows the number of restaurants for each rating
rating = sns.catplot(data = restoX, kind = 'count', x = 'rate', order = restoX['rate'].value_counts().index)
plt.title('Number of restaurants for each rating')
rating.set_xticklabels(rotation = 90)
plt.savefig('/code/output/count_restaurants_based_on_rating.png')

print(restoX['rate'].value_counts())
"""
Analysis 8:
Rating follows Normal distribution and also very few restaurants
has more than 4.5 rating.
"""
print("Analysis 8:")
print("Rating follows Normal distribution and also very few restaurants "
"has more than 4.5 rating.")

# figure shows top 5 restaurants with meal type
meal = sns.catplot(data = restoX, kind = 'count', x = 'rest_type', order = restoX['rest_type'].value_counts().head(5).index)
plt.title('Top 5 Number of restaurants for each type')
meal.set_xticklabels(rotation = 90)
plt.savefig('/code/output/top_restaurants_type_based.png')

print(restoX['rest_type'].value_counts().head(5))
#Analysis 9: Quick Bites, Casual Dining  and Cafes are three top resturent type.
print("Analysis 9:")
print("Quick Bites, Casual Dining  and Cafes are three top resturent type.")


# figure shows bottom 5 restaurants with meal type
me = sns.catplot(data = restoX, kind = 'count', x = 'rest_type', order = restoX['rest_type'].value_counts().tail(5).index)
plt.title('Bottom 5 Number of restaurants for each type')
me.set_xticklabels(rotation = 90)
plt.savefig('/code/output/bottom_restaurants_type_based.png')

print(restoX['rest_type'].value_counts().tail(5))
#Analysis 10: Confectionery, Microbrewery, Food Truck etc. are  bottom resturent type.
print("Analysis 10:")
print("Confectionery, Microbrewery, Food Truck etc. are  bottom resturent type.")


# figure shows top 5 restaurants based on votes
restoX[['votes', 'name']].groupby(['name']).median().sort_values("votes", ascending = False).head(5).plot.bar()
plt.title('Bar plot of votes vs names')
plt.savefig('/code/output/top_voted_restaurants_by_customer.png')

print(restoX[['votes', 'name']].groupby(['name']).median().sort_values("votes", ascending = False).head(5))
#Analysis 11: Top 3 voted restaurants are Prost Brew Pub, Church Street Social and Hoot.
print("Analysis 11:")
print("Top 3 voted restaurants are Prost Brew Pub, Church Street Social and Hoot.")

# removing ',' character from values in approx_cost(for two people)
restoX['approx_cost(for two people)'] = restoX['approx_cost(for two people)'].apply(lambda x: str(x).replace(',',''))
avg = list(restoX['approx_cost(for two people)'])
try:
    for i in range(0, len(avg)):
        if avg[i] != 'nan':
            avg[i] = int(avg[i])
        else:
            avg[i] = None
except:
    pass
restoX['approx_cost(for two people)'] = avg

# figure shows listed_in(city) vs approx_cost(for two people)
restoX[['approx_cost(for two people)', 'listed_in(city)']].groupby(['listed_in(city)']).median().sort_values("approx_cost(for two people)", ascending = False).plot.bar()
plt.title('Bar plot of approx_cost(for two people) vs listed_in(city)')
plt.savefig('/code/output/listed_in(city)_vs_approx_cost(for_two_people).png')

print(restoX[['approx_cost(for two people)', 'listed_in(city)']].groupby(['listed_in(city)']).median().sort_values("approx_cost(for two people)", ascending = False))
#Analysis 12: Church Street is very costly to eat while Banashankari is least costly.
print("Analysis 12: ")
print("Church Street is very costly to eat while Banashankari is least costly.")

# figure shows listed_in(city) vs rate
restoX[['rate', 'listed_in(city)']].groupby(['listed_in(city)']).median().sort_values("rate", ascending = False).plot.bar()
plt.title('Bar plot of rate vs listed_in(city)')
plt.savefig('/code/output/listed_in(city)_vs_rate.png')

print(restoX[['rate', 'listed_in(city)']].groupby(['listed_in(city)']).median().sort_values("rate", ascending = False))
"""
Analysis 13:
Higest rested cities are Brigade Road and Church Street 
while lowest rated city is Brookefield.
"""
print("Analysis 13:")
print("Higest rested cities are Brigade Road and Church Street " 
"while lowest rated city is Brookefield.")

# figure shows top 5 cuisines with maximum number of restaurants
cuis= sns.countplot(data = restoX, x = 'cuisines', order = restoX.cuisines.value_counts().iloc[: 5].index)
cuis.set_xticklabels(cuis.get_xticklabels(),rotation = 90)
plt.title('Top 5 cuisines with maximum number of restaurants')
plt.savefig('/code/output/top_5_cuisines.png')

print(restoX.cuisines.value_counts().iloc[: 5])
#Analysis 14: top cuisines are North Indian and  Chinese.

print("Analysis 14:")
print("top cuisines are North Indian and  Chinese.")
```
